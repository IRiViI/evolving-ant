## Mutants vs clones vs children

```python
dataframe_mutants = dataframe[
    (dataframe["mutated"] == True) &
    (dataframe["cloned"] == True)
]
dataframe["isMutant"] = (
    (dataframe["mutated"] == True) &
    (dataframe["cloned"] == True)
)

dataframe_clones = dataframe[
    (dataframe["mutated"] == False) &
    (dataframe["cloned"] == False)
]
dataframe["isClone"] = (
    (dataframe["mutated"] == False) &
    (dataframe["cloned"] == False)
)

dataframe_children = dataframe[
    (dataframe["mutated"] == True) &
    (dataframe["cloned"] == False)
]
dataframe["isChild"] = (
    (dataframe["mutated"] == True) &
    (dataframe["cloned"] == False)
)
```

```python
mutant_fitness = dataframe_mutants.groupby('episode')['fitness_value']
clones_fitness = dataframe_clones.groupby('episode')['fitness_value']
children_fitness = dataframe_children.groupby('episode')['fitness_value']
```

### Means

```python
plt.figure(figsize=[14,14])
plt.plot(mutant_fitness.mean())
plt.plot(clones_fitness.mean())
plt.plot(children_fitness.mean())
plt.legend(['mutant','clones','children'])
```

```python
plt.figure(figsize=[14,14])
plt.plot(gaussian_filter(mutant_fitness.mean(),10))
plt.plot(gaussian_filter(clones_fitness.mean(),10))
plt.plot(gaussian_filter(children_fitness.mean(),10))
plt.legend(['mutant','clones','children'])
```

### Max

```python
plt.figure(figsize=[14,14])
plt.plot(mutant_fitness.nlargest(5).max())
plt.plot(clones_fitness.max())
plt.plot(children_fitness.max())
plt.legend(['mutant','clones','children'])
```

```python
plt.figure(figsize=[14,14])
plt.plot(gaussian_filter(mutant_fitness.max(),10))
plt.plot(gaussian_filter(clones_fitness.max(),10))
plt.plot(gaussian_filter(children_fitness.max(),10))
plt.legend(['mutant','clones','children'])
```

### Max 5

```python
plt.figure(figsize=[14,14])
plt.plot(mutant_fitness.nlargest(5).mean(level=0))
plt.plot(clones_fitness.nlargest(5).mean(level=0))
plt.plot(children_fitness.nlargest(5).mean(level=0))
plt.legend(['mutant','clones','children'])
```

```python
plt.figure(figsize=[14,14])
plt.plot(gaussian_filter(mutant_fitness.nlargest(5).mean(level=0),10))
plt.plot(gaussian_filter(clones_fitness.nlargest(5).mean(level=0),10))
plt.plot(gaussian_filter(children_fitness.nlargest(5).mean(level=0),10))
plt.legend(['mutant','clones','children'])
```

# Follow 1 lineage

```python
row = dataframe.iloc[-1]
counter = 0
line = []
while row.parent_0:
    line.append(row)
    try:
        row = dataframe[dataframe.id == row.parent_0].iloc[0]
        counter+=1
        if counter > 1000:
            break
    except:
        break
line = pd.DataFrame(line)
```

```python
line.mutated = line.mutated * 1
line.cloned = line.cloned * 1
line.isChild = line.isChild * 1
line.isMutant = line.isMutant * 1
line.isClone = line.isClone * 1
line.describe()
```

What I had             --->           This lineage
- 60% Mutants      --->   43.61% Mutants
- 20% Children     --->   13.50% Children
- 20% Survivors    --->   42.88% Survivors


```python
line.plot.scatter(x='episode',y='fitness_value',figsize=(14,14))
```

```python
line_clones = line[line['isClone']==1]
line_mutants = line[line['isMutant']==1]
line_children = line[line['isChild']==1]
ax = line_clones.plot.scatter(x='episode',y='fitness_value',figsize=(14,14), color='green')
ax = line_mutants.plot.scatter(x='episode',y='fitness_value',figsize=(14,14),ax=ax, color='red')
ax = line_children.plot.scatter(x='episode',y='fitness_value',figsize=(14,14),ax=ax, color='blue')
plt.legend(['Survivor','Mutant','Child'])
```

# Follow 5 Lineages

```python
lines = []
for i in range(5):
    row = dataframe.iloc[-i]
    counter = 0
    line = []
    while row.parent_0:
        line.append(row)
        try:
            row = dataframe[dataframe.id == row.parent_0].iloc[0]
            counter+=1
            if counter > 1000:
                break
        except:
            break
    line = pd.DataFrame(line)
    lines.append(line)
```

```python
plt.figure()
ax = None
for line,color in zip(lines, ['red','green','blue','orange','black']):
#     print(color)
    ax = line.plot.scatter(x='episode',y='fitness_value',figsize=(14,14),color=color,ax=ax)
```
